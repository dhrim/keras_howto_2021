# 순차열 데이터 그리기

# 임포팅

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 대상 데이터

```python
x = np.linspace(0, 10, 100)
s1 = np.sin(x)
s2 = np.cos(x)
s3 = np.sin(x+1)*0.5
```

# 기본 그리기

## 1개 그리기


```python
x = np.linspace(0, 10, 100)

plt.figure(figsize=(8,2))
plt.plot(s1)
plt.show()
```

## 같이 그리기

```python
fig = plt.figure()

ax1 = fig.add_subplot(3,1,1)
ax1.plot(s1)

ax2 = fig.add_subplot(3,1,2)
ax2.plot(s2)

ax3 = fig.add_subplot(3,1,3)
ax3.plot(s3)

plt.show()
```

## 한곳에 같이 그리기

```python
x = np.linspace(0, 10, 100)

plt.figure(figsize=(8,2))
plt.plot(s1)
plt.plot(s2)
plt.plot(s3)

plt.show()
```

# plot() 옵션들

## 라인 색깔
- 가능한 컬러 설정 방법 : https://matplotlib.org/stable/tutorials/colors/colors.html
- 컬러 이름별 컬러 보기 : https://matplotlib.org/stable/gallery/color/named_colors.html

```python
plt.style.use('seaborn-whitegrid')
```

```python
x = np.linspace(0, 5, 100)
```

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

## 라인 스타일

- marker list : https://matplotlib.org/api/markers_api.html

```python
plt.plot(x + 0, linestyle='solid')
plt.plot(x + 1, linestyle='dashed')
plt.plot(x + 2, linestyle='dashdot')
plt.plot(x + 3, linestyle='dotted');

plt.plot(x + 4, linestyle='-')  # solid
plt.plot(x + 5, linestyle='--') # dashed
plt.plot(x + 6, linestyle='-.') # dashdot
plt.plot(x + 7, linestyle=':');  # dotted
```

## 라인 색깔과 스타일 같이

```python
plt.plot(x + 0, '-g')  # solid green
plt.plot(x + 1, '--c') # dashed cyan
plt.plot(x + 2, '-.k') # dashdot black
plt.plot(x + 3, ':r');  # dotted red
```
